```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('./Advertising.csv')
```

```python
df.head()
```

```python
# find correlation
for column in df.columns:
    corr = np.corrcoef(df['sales'],df[column])
    print(f"Correlation betweem Sales and {column} = {corr[0,1]}")
```

```python
x = df.drop(['sales','newspaper'],axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=12345)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train,y_train)
```

```python
LinearRegression()
```

```python
y_predictions = model.predict(x_test)
```

```python
print(y_predictions)
```

```python
plt.scatter(x_test['TV'],y_test)
plt.plot(x_test['TV'],y_predictions,color='red',label="regression line")
plt.legend()
plt.xlabel('TV')
plt.ylabel('sales')
plt.title('TV Vs Sales OBSERVED OUTPUT')
```

```python
plt.scatter(x_test['radio'],y_test)
plt.plot(x_test['radio'],y_predictions,color='red',label="regression line")
plt.legend()
plt.xlabel('radio')
plt.ylabel('sales')
plt.title('radio Vs Sales OBSERVED OUTPUT')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

MAE = mean_absolute_error(y_test,y_predictions)
MSE = mean_squared_error(y_test,y_predictions)
RMSE = np.sqrt(MSE)
print(f"Mean Absoulte Error = {MAE}")
print(f"Mean Sqaured Error = {MSE}")
print(f"Root Mean Squared Error ={RMSE}")
```
